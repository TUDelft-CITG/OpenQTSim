```python
%%time
if __name__ == '__main__': # protect your program's entry point
    
    import pandas as pd
    import numpy as np
    import openqtsim

    from multiprocessing import Pool
    from queueing_engine import worker, Task

    # select nr of servers
    nr_servers = 10
    servers =  np.array(range(1, nr_servers + 1))

    # labda is selected to be ranging from .1 to .9
    lams = np.array(range(1,10))

    # mu is selected depending on the nr of servers to always be 10 
    mu = 10/nr_servers

    # lambda divided by (mu * nr_servers) provides the utilisations
    utilisations = lams / (mu * nr_servers)

    # every instance will be run with nr_arr number of customers
    nr_arr = 500

    mu=10

    # # initialise dataframe
    # df = pd.DataFrame(index=utilisations, columns=servers)

    # tasks = [Task(1, 2),Task(3, 4)]
    tasks=[]
    # one by one fill the dataframe
    for index, lam in enumerate(lams):
        for server in servers:
            tasks.append(Task(lam, mu, server, nr_arr))

    p = Pool(5)
    result = p.map(worker, tasks)
    p.close()
```

```python
print('MT', result)
```

```python
df = pd.DataFrame(tasks)
df['a'] = result
df['utilisation'] = df.lam / df.mu
```
