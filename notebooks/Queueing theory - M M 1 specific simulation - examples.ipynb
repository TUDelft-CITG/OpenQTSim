# Queueing theory - M/M/1 specific simulation - examples

### Simple test 

```python
# simple example for testing
import openqtsim
import pandas as pd

IAT=[2,2,2,2,2,2,2,2]
ST= [3,1,3,1,3,1,3,1]

df = pd.DataFrame()
df["IAT"]=IAT
df["ST"]=ST

MM1 = openqtsim.MM1(IAT, ST, len(IAT))
df = MM1.calculate_MM1(IAT, ST)
MM1.get_stats(df)
df
```

### Example of extracting IAT and ST data from Excel file

```python
import openqtsim
import pandas as pd
import xlrd

df = pd.read_excel('Queueing theory - D D 1 example.xlsx', index_col=None, header=0, sheet_name='data')
IAT = df["IAT"]
ST = df["ST"]

MM1 = openqtsim.MM1(IAT, ST, len(IAT))
df = MM1.calculate_MM1(IAT, ST)
MM1.get_stats(df)
df
```
